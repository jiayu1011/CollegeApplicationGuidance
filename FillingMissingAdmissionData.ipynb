## 根据“一分一段”填充缺失的数据

```python
import numpy as np
import pandas as pd
import os
```

```python
os.getcwd()
```

```python
scoreRank2018DataPath = './data/2018广东省高考一分一段数据.txt'
scoreRank2019DataPath = './data/2019广东省高考一分一段数据.txt'
scoreRank2020DataPath = './data/2020广东省高考一分一段数据.txt'
```

```python
scoreRank2018Data = {}
scoreRank2019Data = {}
scoreRank2020Data = {}

with open(scoreRank2018DataPath, 'r', encoding='utf-16') as f2018:
    linecount = 0
    columns = []
    while True:
        line = f2018.readline()
        linecount += 1
        if not line:
            break
        
        rowDataArr = line.strip().split('\t')

        if linecount==1:    
            columns = rowDataArr
            for (index, item) in enumerate(columns):
                scoreRank2018Data[item] = []   
        elif linecount>1:
            
            if not line.strip():
                continue
            for (index, item) in enumerate(rowDataArr):
                scoreRank2018Data[columns[index]].append(item)   
                
                
with open(scoreRank2019DataPath, 'r', encoding='utf-16') as f2019:
    linecount = 0
    columns = []
    while True:
        line = f2019.readline()
        linecount += 1
        if not line:
            break
        
        rowDataArr = line.strip().split('\t')

        if linecount==2:    
            columns = rowDataArr
            for (index, item) in enumerate(columns):
                scoreRank2019Data[item] = []   
        elif linecount>2:
            
            if not line.strip():
                continue
            for (index, item) in enumerate(rowDataArr):
                scoreRank2019Data[columns[index]].append(item)      
                
                
with open(scoreRank2020DataPath, 'r', encoding='utf-16') as f2020:
    linecount = 0
    columns = []
    while True:
        line = f2020.readline()
        linecount += 1
        if not line:
            break
        
        rowDataArr = line.strip().split('\t')

        if linecount==2:    
            columns = rowDataArr
            for (index, item) in enumerate(columns):
                scoreRank2020Data[item] = []   
        elif linecount>2:
            
            if not line.strip():
                continue
            for (index, item) in enumerate(rowDataArr):
                scoreRank2020Data[columns[index]].append(item)                   
    
```

```python
df_scoreRank2018 = pd.DataFrame(scoreRank2018Data)
df_scoreRank2019 = pd.DataFrame(scoreRank2019Data)
df_scoreRank2020 = pd.DataFrame(scoreRank2020Data)
```

```python
df_scoreRank2018
```

```python
scoreRankDict = {
    '2018': df_scoreRank2018,
    '2019': df_scoreRank2019,
    '2020': df_scoreRank2020
}
```

## 导入录取数据

```python
admissionDataPath = './data/中山大学2018-2020年本科批次录取统计（广东）.txt'
```

```python
admissionData = {}  
with open(admissionDataPath, 'r', encoding='utf-16') as f:
    linecount = 0
    columns = []  
    while True:
        line = f.readline()
        if not line:
            break
        linecount += 1
        
        if linecount==1:
            columns = line.strip().split('\t')
#             print(columns)
            for (index, item) in enumerate(columns):
                admissionData[item] = []
            
        if line.strip() and linecount!=1:
            rowDataArr = line.strip().split('\t')
            # 在缺失的数据中补上''作为填充
            while len(rowDataArr)<len(columns):
                rowDataArr.append('')
                
            for (index, item) in enumerate(rowDataArr):
                admissionData[columns[index]].append(item)
```

```python
df_admission = pd.DataFrame(admissionData)
```

```python
df_admission
```

```python
columns_years = ['2020最低分', '2020平均分', '2019最低分', '2019平均分', '2018最低分', '2018平均分']
years = list(map(lambda x: x[:4], columns_years))
appendRankDataDict = {
    '2020最低分排位': [],
    '2020平均分排位': [],
    '2019最低分排位': [],
    '2019平均分排位': [],
    '2018最低分排位': [],
    '2018平均分排位': []
}
for (index_column, column) in df_admission.iteritems():
    if not index_column in columns_years:
        continue

    for (index_score, score) in enumerate(column):
        if not score:
            appendRankDataDict[index_column+'排位'].append('')
            continue
        tempDf = scoreRankDict[index_column[:4]]
        scoreRank = tempDf[tempDf['分数段']==score]['累计人数'].tolist()[0]
        appendRankDataDict[index_column+'排位'].append(scoreRank)
```

```python
appendRankDataDict
```

```python
# 插入录取大表
df_admission['2020最低分排位'] = appendRankDataDict['2020最低分排位']
df_admission['2020平均分排位'] = appendRankDataDict['2020平均分排位']
df_admission['2019最低分排位'] = appendRankDataDict['2019最低分排位']
df_admission['2019平均分排位'] = appendRankDataDict['2019平均分排位']
df_admission['2018最低分排位'] = appendRankDataDict['2018最低分排位']
df_admission['2018平均分排位'] = appendRankDataDict['2018平均分排位']
```

```python
# 让排名跟分数挨在一起
df_admission = df_admission.reindex(columns=[
    '专业名称', '所属学院', '科类', 
    '2020最低分', '2020最低分排位',
    '2020平均分', '2020平均分排位',
    '2019最低分', '2019最低分排位',
    '2019平均分', '2019平均分排位',
    '2018最低分', '2018最低分排位',
    '2018平均分', '2018平均分排位'
])
```
