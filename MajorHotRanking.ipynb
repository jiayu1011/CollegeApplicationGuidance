# 中山大学文科/理科 专业热度排行（近三年）

```python
import numpy as np
import pandas as pd
import os
```

#### 注: 部分数据存在缺失,故计算热门/冷门排行时只使用有数据的专业进行统计

```python
os.getcwd()
```

```python
file_path = "./data/中山大学2018-2020年本科批次录取统计（广东）.txt"
```

```python
data = {}  
with open(file_path, 'r', encoding='utf-16') as f1:
    linecount = 0
    columns = []  
    while True:
        line = f1.readline()
        if not line:
            break
        linecount += 1
        
        if linecount==1:
            columns = line.strip().split('\t')
#             print(columns)
            for (index, item) in enumerate(columns):
                data[item] = []
            
        if line.strip() and linecount!=1:
            rowDataArr = line.strip().split('\t')
            # 在缺失的数据中补上''作为填充
            while len(rowDataArr)<len(columns):
                rowDataArr.append('')
                
            for (index, item) in enumerate(rowDataArr):
                data[columns[index]].append(item)
        
    
            
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
# 共117条数据,部分数据存在缺失,故计算热门/冷门排行时只使用有数据的专业进行统计
df
```

```python
# 统计最热门/最冷门的十大专业
data_wen_HotAndCold = {}
data_li_HotAndCold = {}

df_wen = df[df['科类']=='文科'][['专业名称', '科类',  '2020平均分', '2019平均分', '2018平均分']]
df_li = df[df['科类']=='理科'][['专业名称', '科类',  '2020平均分', '2019平均分', '2018平均分']]

# 文科
# 降序排列
df_wen.sort_values('2020平均分', inplace=True, ascending=False)
data_wen_HotAndCold['2020年热门专业排行'] = df_wen[df_wen['2020平均分']!='']['专业名称'][:10].tolist()
# 升序排列
df_wen.sort_values('2020平均分', inplace=True, ascending=True)
data_wen_HotAndCold['2020年冷门专业排行'] = df_wen[df_wen['2020平均分']!='']['专业名称'][:10].tolist()

df_wen.sort_values('2019平均分', inplace=True, ascending=False)
data_wen_HotAndCold['2019年热门专业排行'] = df_wen[df_wen['2019平均分']!='']['专业名称'][:10].tolist()
df_wen.sort_values('2019平均分', inplace=True, ascending=True)
data_wen_HotAndCold['2019年冷门专业排行'] = df_wen[df_wen['2019平均分']!='']['专业名称'][:10].tolist()
df_wen.sort_values('2018平均分', inplace=True, ascending=False)
data_wen_HotAndCold['2018年热门专业排行'] = df_wen[df_wen['2018平均分']!='']['专业名称'][:10].tolist()
df_wen.sort_values('2018平均分', inplace=True, ascending=True)
data_wen_HotAndCold['2018年冷门专业排行'] = df_wen[df_wen['2018平均分']!='']['专业名称'][:10].tolist()


# 理科
# 降序排列
df_li.sort_values('2020平均分', inplace=True, ascending=False)
data_li_HotAndCold['2020年热门专业排行'] = df_li[df_li['2020平均分']!='']['专业名称'][:10].tolist()
# 升序排列
df_li.sort_values('2020平均分', inplace=True, ascending=True)
data_li_HotAndCold['2020年冷门专业排行'] = df_li[df_li['2020平均分']!='']['专业名称'][:10].tolist()

df_li.sort_values('2019平均分', inplace=True, ascending=False)
data_li_HotAndCold['2019年热门专业排行'] = df_li[df_li['2019平均分']!='']['专业名称'][:10].tolist()
df_li.sort_values('2019平均分', inplace=True, ascending=True)
data_li_HotAndCold['2019年冷门专业排行'] = df_li[df_li['2019平均分']!='']['专业名称'][:10].tolist()
df_li.sort_values('2018平均分', inplace=True, ascending=False)
data_li_HotAndCold['2018年热门专业排行'] = df_li[df_li['2018平均分']!='']['专业名称'][:10].tolist()
df_li.sort_values('2018平均分', inplace=True, ascending=True)
data_li_HotAndCold['2018年冷门专业排行'] = df_li[df_li['2018平均分']!='']['专业名称'][:10].tolist()
```

```python
# 使第一个索引保持为1
df_wen_HotAndCold = pd.DataFrame(data_wen_HotAndCold)
df_li_HotAndCold = pd.DataFrame(data_li_HotAndCold)

df_wen_HotAndCold.index = df_wen_HotAndCold.index + 1
df_li_HotAndCold.index = df_li_HotAndCold.index + 1
```
